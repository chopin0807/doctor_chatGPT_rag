```python
# 본내용에서 '말머리'부분 삭제
df['본내용'] = df['본내용'].apply(lambda x: str(x).replace("* 말머리 선택체크/공지등록 체크금지(운영진만 가능)* 하단 태그 단어표기/성격에 맞는 게시판 등록* 병원(선생님)실명 비방 및 상업적 홍보 금지/병원추천 신중* 회원간 상호비방 금지 및 예절준수https://cafe.naver.com/aclknee/141278무릎관련 협력병원, 재활센터 및 관련용품-http://cafe.naver.com/aclknee/5975카페 필독 내용-https://cafe.naver.com/aclknee/25670카페 운영규칙-https://cafe.naver.com/aclknee/99335병원홍보, 바이럴마켓팅금지-http://cafe.naver.com/aclknee/7752http://db.blueweb.co.kr/formmail/formmail.html?dataname=eximkim0국가유공,후유장해,보상,사고합의 전문가상담-https://band.us/band/64144732밴드모임", ""))
df = df[['본문제목', '본내용', '댓글']]
df
```

```python
# 본문제목, 본내용, 댓글에서 교수 및 의사에 대한 내용 존재하는지 확인 => 각각 키워드(본문제목), 키워드(본내용), 키워드(댓글)
# 내용 존재시 해당 내용 리스트로 표시
import re

reg = "[ㄱ-ㅎ]+원장|[ㄱ-ㅎ]+의사|[ㄱ-ㅎ]+대표원장|[ㄱ-ㅎ]+선생|[ㄱ-ㅎ]+쌤|[ㄱ-ㅎ]+교수|[ㄱ-ㅎ]+[ ]원장|[ㄱ-ㅎ]+[ ]의사|[ㄱ-ㅎ]+[ ]대표원장|[ㄱ-ㅎ]+[ ]선생|[ㄱ-ㅎ]+[ ]쌤|[ㄱ-ㅎ]+[ ]교수"
keyword = lambda x: "".join(re.findall(reg, str(x))[i] if i == 0 else "," + re.findall(reg, str(x))[i] for i in range(len(re.findall(reg, str(x)))))
df["키워드(본문제목)"] = df["본문제목"].apply(keyword)
df["키워드(본내용)"] = df["본내용"].apply(keyword)
df["키워드(댓글)"] = df["댓글"].apply(keyword)
df
```

```python
# 본문제목, 본내용, 댓글에서 교수, 원장, 의사가 언급된 부분만 추출
df["본내용"] = df["본내용"].apply(lambda x: str(x).replace("제목 : 25글자 이내로 간략하게 내용 :",""))
df["키워드(본문제목)"] = df["키워드(본문제목)"].apply(lambda x: x if x else "X")
df["키워드(본내용)"] = df["키워드(본내용)"].apply(lambda x: x if x else "X")
df["키워드(댓글)"] = df["키워드(댓글)"].apply(lambda x: x if x else "X")
df = df[(df["키워드(본문제목)"] != "X") | (df["키워드(본내용)"] != "X") | (df["키워드(댓글)"] != "X")]
df
```

```python
df = df.reset_index(drop = True)
df
```

```python
# 의사 및 교수별로 그룹핑
# 본문제목
title = []
for i in df["키워드(본문제목)"]:
    if i != 'X' and i.replace(" ", "") not in title:
        title.append(i.replace(" ", ""))
title
```

```python
title_value = [] # 실제 본문제목 내용
for i in title:
    target = ''
    for j, k in zip(df["키워드(본문제목)"], df["본문제목"]):
        if i in j.replace(" ", ""):
            target += k
            target += '\n'
    reg = re.findall(" [0-9가-힣ㄱ-ㅎ !?~@#$%^&*()\[\],.:;\"']+(?=\\n)", target)
    title_value.append(''.join(i.replace(j, "") for i in reg for j in re.findall("( \[ \d+ \])", i)) if ''.join(i.replace(j, "") for i in reg for j in re.findall("( \[ \d+ \])", i)) else ''.join(i for i in reg))
title_value
```

```python
len(title) == len(title_value) # 그룹화를 위해 두 항목의 길이가 같은지 확인
```

```python
# 본내용
context = []
for i in df["키워드(본내용)"]:
    if i != 'X':
        paste = i.replace(" ", "").split(",")
        for j in paste:
            if j not in context:
                context.append(j)
context
```

```python
context_value = [] # 실제 본내용
for i in context:
    target = ''
    for j, k in zip(df["키워드(본내용)"], df["본내용"]):
        j = j.replace(" ", "").split(",")
        if i in j:
            target += k
    reg = re.findall(" [0-9가-힣ㄱ-ㅎ !?~@#$%^&*()\[\],.:;\"']+(?=\\n)", target)
    context_value.append(''.join(i.replace(j, "") for i in reg for j in re.findall("( \[ \d+ \])", i)) if ''.join(i.replace(j, "") for i in reg for j in re.findall("( \[ \d+ \])", i)) else target.replace(" 손상된 부위, 현상황 , 궁금증 등 기재(성격에 맞는 게시판 및 말머리 체크) ", "").replace("\u200b", "").replace("/n", ""))
context_value
```

```python
len(context) == len(context_value) # 그룹화를 위해 두 항목의 길이가 같은지 확인
```

```python
# 댓글
review = []
for i in df["키워드(댓글)"]:
    if i != 'X':
        paste = i.replace(" ", "").split(",")
        for j in paste:
            if j not in review:
                review.append(j)
review
```

```python
review_value = [] # 실제 댓글
for i in review:
    target = ''
    for j, k in zip(df["키워드(댓글)"], df["댓글"]):
        j = j.replace(" ", "").split(",")
        if i in j:
            target += k
    reg = re.findall(" [0-9가-힣ㄱ-ㅎ !?~@#$%^&*()\[\],.:;\"']+(?=\\n)", target)
    review_value.append(''.join(i.replace(j, "") for i in reg for j in re.findall("( \[ \d+ \])", i)) if ''.join(i.replace(j, "") for i in reg for j in re.findall("( \[ \d+ \])", i)) else target)
review_value
```

```python
len(review) == len(review_value) # 그룹화를 위해 두 항목의 길이가 같은지 확인
```

```python
# 본문제목, 본내용, 댓글에서 언급한 의사, 교수에 대하여 하나의 리스트로 정리 후 사전 순 정렬
doctor = []
for i in title:
    if i not in doctor:
        doctor.append(i)
for j in context:
    if j not in doctor:
        doctor.append(j)
for k in review:
    if k not in doctor:
        doctor.append(k)

doctor.sort()
doctor
```

```python
# 각 언급 의사 및 교수별 본문제목, 본내용, 댓글 통합
doctor_title, doctor_context, doctor_review = [], [], []
for i in doctor: # 언급 의사별 통합 작업
    doctor_title.append(title_value[title.index(i)] if i in title else 'X')
    doctor_context.append(context_value[context.index(i)] if i in context else 'X')
    doctor_review.append(review_value[review.index(i)] if i in review else 'X')

total = pd.DataFrame({'언급 의사/교수':doctor, '본문제목':doctor_title, '본내용':doctor_context, '댓글':doctor_review})
total
```

```python
# csv 파일 작성
total.to_csv("그룹핑 결과.csv", encoding = "utf-8")
```

```python
# chatGPT rag 방식을 위한 text파일 작성
f = open("TextRAG.txt", "w", encoding = "utf-8")
for a, b, c, d in zip(total["언급 의사/교수"], total["본문제목"], total["본내용"], total["댓글"]):
    target = a + " "
    if b != "X":
        target += b
    if c != "X":
        target += c
    if d != "X":
        target += d
    target += "\n"
    f.write(target)
f.close()
```
